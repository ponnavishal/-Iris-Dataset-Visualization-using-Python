```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
# üìå Step 4: Display the first few rows of the dataset
print("üîç First 5 Rows of the Iris Dataset:")
print(df.head())
```

```python
sns.pairplot(df, hue="species", palette="Set2")
plt.suptitle(" Pairplot: Feature Relationships by Species", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("üî• Heatmap: Feature Correlation")
plt.show()
```

```python
# Sepal length
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df, palette="Set3")
plt.title("üì¶ Boxplot: Sepal Length per Species")
plt.show()
```

```python
# Petal width
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal width (cm)', data=df, palette="Set2")
plt.title("üì¶ Boxplot: Petal Width per Species")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal length (cm)', data=df, palette="Pastel1")
plt.title("üéª Violin Plot: Petal Length Distribution by Species")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for species in df['species'].unique():
    sns.histplot(df[df['species'] == species]['petal length (cm)'], label=species, kde=True)
plt.legend()
plt.title("üìä Histogram: Petal Length by Species")
plt.xlabel("Petal Length (cm)")
plt.show()
```
